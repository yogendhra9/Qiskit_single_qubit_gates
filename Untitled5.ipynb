```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

# Initialize simulator
sim = AerSimulator()

# Create a single-qubit circuit with one classical bit
qc = QuantumCircuit(1, 1)

# --- Apply sequence of single-qubit gates ---
qc.x(0)        # Pauli-X (NOT gate)
qc.h(0)        # Hadamard (superposition)
qc.s(0)        # Phase gate (π/2)
qc.t(0)        # T gate (π/4)
qc.rz(0.5, 0)  # Z-axis rotation by 0.5 radians

# Measure final state
qc.measure_all()

# Display circuit
print("Quantum Circuit:")
print(qc.draw())

# --- Run simulation ---
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()

# --- Display results ---
counts = result.get_counts()
print("\nMeasurement Counts:", counts)
```

```python
# --- Task 1: Swap X and H gates ---
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

sim = AerSimulator()

# Create 1-qubit, 1-classical bit circuit
qc = QuantumCircuit(1, 1)

# Modified gate order: H before X
qc.h(0)
qc.x(0)
qc.s(0)
qc.t(0)
qc.rz(0.5, 0)

# Measure
qc.measure_all()

# Display
print("Quantum Circuit (H before X):")
print(qc.draw())

# Run simulation
compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()
counts = result.get_counts()
print("\nMeasurement Counts:", counts)
```

```python
# --- Task 2: Vary RZ rotation angle ---
import math
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

sim = AerSimulator()

angles = [math.pi/4, math.pi/2, math.pi]  # θ = π/4, π/2, π

for theta in angles:
    qc = QuantumCircuit(1, 1)
    qc.x(0)
    qc.h(0)
    qc.s(0)
    qc.t(0)
    qc.rz(theta, 0)
    qc.measure_all()

    print(f"\n--- RZ rotation: θ = {theta:.3f} radians ---")
    print(qc.draw())

    compiled = transpile(qc, sim)
    result = sim.run(compiled, shots=1024).result()
    counts = result.get_counts()
    print("Measurement Counts:", counts)
```

```python
# --- Task 3: Remove Hadamard gate ---
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

sim = AerSimulator()

qc = QuantumCircuit(1, 1)

# Original: X → H → S → T → RZ(0.5)
# Modified: X → S → T → RZ(0.5)
qc.x(0)
qc.s(0)
qc.t(0)
qc.rz(0.5, 0)

# Measure
qc.measure_all()

# Display
print("Quantum Circuit (without Hadamard):")
print(qc.draw())

compiled = transpile(qc, sim)
result = sim.run(compiled, shots=1024).result()
counts = result.get_counts()
print("\nMeasurement Counts:", counts)
```

```python
#-------------------Visualization-------
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

# Example for any circuit 'qc'
# Create a copy of the circuit and remove measurement for visualization
qc_no_measure = qc.remove_final_measurements(inplace=False)

state = Statevector.from_instruction(qc_no_measure)

# Print the statevector to confirm it's generated
print("Statevector:", state)

fig = plot_bloch_multivector(state)

# Save the plot to a file and display it
fig.savefig("bloch_sphere.png")
from IPython.display import Image
display(Image("bloch_sphere.png"))

plt.show()
```
